```python
# Removing the Territories and Dsitrict of Columbia

territoryList = []

for line in state_data.itertuples():
    check_name = str(state_data.at[line.Index, 'STATE'])
    if ('Territory' in check_name):
        state_data.drop(index=line.Index, inplace=True)
        
for line in state_data.itertuples():
    if (str(state_data.at[line.Index, 'STATE']) == 'District Of Columbia'):
        state_data.drop(index=line.Index, inplace=True)

print(state_data['STATE'].tolist())
```

```python
#Adding Slavery/non Slavery and CSA/USA fields 

USA = ['California', 'Connecticut', 'Delaware', 'Illinois', 'Indiana', 'Iowa', 'Kentucky', 'Maine', 
       'Maryland', 'Massachusetts', 'Michigan', 'Minnesota', 'Missouri', 'New Hampshire', 'New Jersey',
       'New York', 'Ohio', 'Oregon', 'Pennsylvania', 'Rhode Island', 'Vermont', 'Wisconsin']

CSA = ['Alabama', 'Arkansas', 'Florida', 'Georgia','Louisiana', 'Mississippi', 'North Carolina', 
       'South Carolina', 'Tennessee', 'Texas', 'Virginia']

SLAV = ['Alabama', 'Arkansas', 'Florida', 'Georgia','Louisiana', 'Mississippi', 'North Carolina', 
       'South Carolina', 'Tennessee', 'Texas', 'Virginia', 'Delaware', 'Kentucky', 'Maryland', 'Missouri']

sepList = []
slaveList = []

#This is a bit more complicated than necessary to ensure that no state is missed
for line in state_data.itertuples():
    
    if (state_data.at[line.Index, 'STATE'] in USA):
        sepList.append('USA')
    elif (state_data.at[line.Index,'STATE'] in CSA):
        sepList.append('CSA')
    else:
        print('error')
 
    if (state_data.at[line.Index, 'STATE'] in SLAV):
        slaveList.append("Slave")
    else:
        slaveList.append("Free")

state_data['Support'] = sepList
state_data['Slavery'] = slaveList

# Need to reset the index to account for deleted rows
state_data.reset_index(drop=True, inplace=True)

state_data.head()
```
