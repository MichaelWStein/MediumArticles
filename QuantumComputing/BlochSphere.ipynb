```python
from math import sqrt, pi
from qiskit.visualization import plot_histogram, plot_bloch_vector
```

```python
#This is needed for the simulation at the end
from qiskit import QuantumCircuit, Aer
from qiskit.visualization import plot_bloch_multivector, plot_histogram
```

```python
#Choosing the simulator
sim = Aer.get_backend('aer_simulator')
```

```python
#Different way of showing the vector - The Bloch vector is for visualization,
#maps the 2D, complex statevector onto real, 3D spaces (IBM qiskit textbook)
#Showing the |0> vector.
from qiskit_textbook.widgets import plot_bloch_vector_spherical
coords = [0,0,1] # [Theta, Phi, Radius]
plot_bloch_vector_spherical(coords) # Bloch Vector with spherical coordinates

#In the long run it may be better not use the textbook program (not sure how stable the maintenance is),
#but use the Qiskit function plot_bloch_vector(), but that uses only Cartesian coordinates
#The calculations in the textbook, however, are for the spherical coordinates
```

```python
#Showing the |1> vector in x, y, z coordinates.

from qiskit.visualization import plot_bloch_vector
%matplotlib inline

plot_bloch_vector([0,0,-1], title="New Bloch Sphere")
```

```python
#Showing the |1> vector.
coords = [pi,0,1] # [Theta, Phi, Radius]
plot_bloch_vector_spherical(coords) # Bloch Vector with spherical coordinates
```

```python
#Bloch Sphere after Hadamard Gate
from qiskit_textbook.widgets import plot_bloch_vector_spherical
coords = [pi/2,0,1] # [Theta, Phi, Radius]
plot_bloch_vector_spherical(coords) # Bloch Vector with spherical coordinates
```

```python
#Easier way to create the Bloch sphere - directly from the statevector
#In this case for a Hadamard Gate

qc = QuantumCircuit(1)
qc.h(0)
qc.draw()
```

```python
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
plot_bloch_multivector(state)
```

```python
#Bloch sphere for a two qubit system
qc = QuantumCircuit(2)
qc.h(0)
qc.x(1)
qc.draw()
```

```python
qc.save_statevector()
state = sim.run(qc).result().get_statevector()
plot_bloch_multivector(state)
```
